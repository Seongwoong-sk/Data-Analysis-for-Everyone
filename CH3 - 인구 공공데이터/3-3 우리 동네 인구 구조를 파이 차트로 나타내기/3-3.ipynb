# 1. 제주도에는 여성의 비율이 높을까

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input("찾고 싶은 지역을 입력해주세요 : ")

for row in data:
    if name in row[0]:
        
        for i in row[3:104] : # 남성인구 인덱스 3 ~ 103
            m.append(-int(i))
        
        for i in row[106:] : # 여성인구 인덱스 106~
            f.append(int(i))
            
        break # name에 해당하는 값을 찾았다면 반복문을 중단해라

# f.reverse() # 리스트의 값을 역순으로 재배열하는 함수


plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  # - 기호 깨짐 방지
plt.title( name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label = '남성')
plt.barh(range(101),f, label = '여성')
plt.legend()
plt.show()
```

# 2. 혈액형 비율 표현하기

### pie() 함수
- 전체 데이터 중 특정 데이터의 비율을 보기 쉽게 표현

```python
import matplotlib.pyplot as plt

plt.pie([10,20])
plt.show()
```

### 동그란 원 그리기 

```python
size = [2441, 2312, 1031, 1233]
plt.axis('equal')
plt.pie(size)
plt.show()
```

### 레이블 추가하기

```python
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ['A형', 'B형', 'AB형', 'O형'] # 레이블

plt.axis('equal')
plt.pie(size, labels = label)
plt.show()
```

### 비율 및 범례 표시하기
- pie() 함수의 autopct 속성은 auto percent를 의미하며, 어떤 형태로 값을 표시할지 작성하면 각 항목의 비율을 자동으로 계산하여 표시함.

```python
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ['A형', 'B형', 'AB형', 'O형'] # 레이블

plt.axis('equal')
plt.pie(size, labels = label, autopct ='%.1f%%')
plt.legend()
plt.show()
```

### 색 및 돌출 효과 정하기

```python
plt.rc('font', family = 'Malgun Gothic') # 그래프에 한글 표시
size = [2441, 2312, 1031, 1233] # 데이터
label = ['A형', 'B형', 'AB형', 'O형'] # 레이블
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']

plt.axis('equal')
plt.pie(size, labels = label, autopct ='%.1f%%' , colors=color, explode=(0,0,0.2,0)) # explode = 특정 칸이 돌출하는 효과
plt.legend()
plt.show()
```

# 3. 제주도의 성별 인구 비율 표현하기

```python
# 성별 인구의 합을 구해야 됨.

import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)
size = []   # 성별 인구 합계 저장할 리스트

name = input("찾고 싶은 지역을 입력해주세요 : ")

for row in data:
    if name in row[0]:
        m = 0
        f = 0
        
        for i in range(101) : 
            m += int(row[i+3])   # 3 ~ 103번까지 데이터 더하기
            f += int(row[i+106]) # 106 ~ 206번까지 데이터 더하기
            
        break # name에 해당하는 값을 찾았다면 반복문을 중단해라

size.append(m)
size.append(f)

# 파이 차트로 표현

plt.rc('font', family = 'Malgun Gothic') 
color = ['crimson' , 'darkcyan']
plt.axis('equal')
                                                    # startangle : 파이 차트의 시작 각도를 정해줌
                                                    # 속성 값을 90으로 하면 반시계방향으로 90도 이동하여 12시 정각위치에서 그래프가 시작
                                                    # 파이 차트 기본 시작 시점 : 3시
plt.pie(size, labels = ['남','여'], autopct ='%.1f%%' , colors=color, startangle = 90)
plt.title(name +' 지역의 남녀 성별 비율')
plt.show()
```
