# 입력 받은 지역의 인구 구조를 파악하자

```python
import csv

f = open('age.csv')
data = csv.reader(f)
result = []
name = input('인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
                          
                          
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rc('font', family = "Malgun Gothic")
plt.title(name + '지역의 인구 구조')
plt.plot(result)
plt.show()
```

# 2. 항아리 모양 그래프 그리기

#### 데이터 수집하기
- 행정안전부 인구 데이터

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

for row in data:
    if '신도림' in row[0]:
        for i in row[3:104] : # 남성인구 인덱스 3 ~ 103
            m.append(-int(i))
        for i in row[106:] : # 여성인구 인덱스 106~
            f.append(int(i))
        

f.reverse() # 리스트의 값을 역순으로 재배열하는 함수

```

#### 데이터 시각화하기

```python
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  # - 기호 깨짐 방지
plt.title('신도림 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m)
plt.barh(range(101),f)
plt.legend()
plt.show()
```

* 그래프가 겹쳐지는 이유 : 남성 , 여성 데이터 모두 양수로 이루어짐 -> 하나를 음수로 변경

### 우리 동네 인구 구조를 항아리 모양 그래프로 그리기

```python
import csv

f = open('gender.csv')
data = csv.reader(f)
m = []
f = []

name = input("찾고 싶은 지역을 입력해주세요 : ")

for row in data:
    if name in row[0]:
        for i in row[3:104] : # 남성인구 인덱스 3 ~ 103
            m.append(-int(i))
        for i in row[106:] : # 여성인구 인덱스 106~
            f.append(int(i))
        

f.reverse() # 리스트의 값을 역순으로 재배열하는 함수


plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False  # - 기호 깨짐 방지
plt.title( name+' 지역의 남녀 성별 인구 분포')
plt.barh(range(101),m, label = '남성')
plt.barh(range(101),f, label = '여성')
plt.legend()
plt.show()
```
