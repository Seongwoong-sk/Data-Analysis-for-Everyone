# 1. 꺾은선 그래프로 표현하기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = [] # 연령별 남성 인구 숫자를 저장할 리스트 생성 
f = [] # 연령별 여성 인구를 저장할 리스트 생성

name = input("궁금한 동네를 입력해주세요 : ")
for row in data:
    if name in row[0] : 
        for i in range(3,104):
            m.append(int(row[i])) # 남성 데이터 저장하기
            f.append(int(row[i+103])) # 여성 데이터 저장하기
        break

import matplotlib.pyplot as plt
plt.plot(m, label = 'Male')
plt.plot(f, label = 'Female')
plt.legend()
plt.show()
```

# 3. 버블 차트로 표현하기
- 버블 차트를 그릴 때도 사용할 수 있음
- 기존 코드에서 size를 의미하는 s 속성을 추가하고 원하는 크기를 입력한다
- c 속성을 추가하면 각 점의 색상도 정할 수 있다
- c 속성 :  표현하고 싶은 색상의 수

```python
plt.scatter([1,2,3,4],[10,30,20,40], s=[100,200,250,300], c= ['red','blue','green','gold'])
plt.style.use('ggplot')

plt.show()
```

#### colorbar() 함수 사용

```python
plt.scatter([1,2,3,4],[10,30,20,40], s=[100,200,250,300], c= range(4), cmap='jet')  # 4개의 점을 각각 다른 색으로 표현한다.
plt.style.use('ggplot')                                                 # cmap 속성은 컬러바에 사용될 색상의 종류를 정할 수 있음.
plt.colorbar()
plt.show()
```

### 랜덤하게 100개 점 산점도로 표현하기

```python
import random

x = []
y = []
size = []
for i in range(100) :
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))    
    size.append(random.randint(10,100))

                            # 컬러바 c 속서에 size 리스트를 사용하면 크기에 따라 다른 색 표현 가능
plt.scatter(x, y, s = size, c =size, cmap = 'brg',alpha = 0.7) # alpha 속성 : 투명도 
                                                                # (범위 : 0 ~ 1, 0에 가까울수록 투명, 1에 가까울수록 반대)
plt.colorbar()
plt.show()
```

# 4. 제주도의 연령대별 성별 비율을 산점도로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv')
data = csv.reader(f)

m = [] # 연령별 남성 인구 숫자를 저장할 리스트 생성 
f = [] # 연령별 여성 인구를 저장할 리스트 생성

name = input("궁금한 동네를 입력해주세요 : ")
for row in data:
    if name in row[0] : 
        for i in range(3,104):
            m.append(int(row[i])) # 남성 데이터 저장하기
            f.append(int(row[i+103])) # 여성 데이터 저장하기
        break

plt.rc('font', family = "Malgun Gothic")
plt.scatter(m, f, c = range(101), alpha = 0.5, cmap='jet') # 컬러맵 적용
plt.colorbar()
plt.plot(range(max(m)), range(max(m)), 'g') # 추세선 추가
plt.xlabel('남성 인구수') # x축 라벨
plt.ylabel('여성 인구수') # y축 라벨
plt.show()
```
