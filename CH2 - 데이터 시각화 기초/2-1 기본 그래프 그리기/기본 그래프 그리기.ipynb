# 1. matplotlib 라이브러리란?

- 파이썬으로 데이터를 시각화하는 데 많이 쓰이는 라이브러리
- 과학 컴퓨터 연구 분야나 인공지능 연구 분야에서 많이 활용됨

### matplotlib 라이브러리 안에 있는 다양한 모듈중 pyplot이라는 모듈

```python
# pyplot 라이브러리 불러오기
import matplotlib.pyplot as plt # plt라는 별명(알리아스) 지정
```

# 2. 기본 그래프 그리기 
plot( ) 함수 - 직선 또는 꺾은선 형태의 그래프를 그릴 때 사용할 수 있는 명령어

```python
# plot( ) 함수에 한 개의 리스트 입력

import matplotlib.pyplot as plt
plt.plot([10,20,30,40])  # matplotlib.pyplot.plot()과 같은 뜻
plt.show() # 그래프로 표현할 때 쓰는 명령어
```

### plot( ) 함수에 입력된 리스트의 값이 y축 값으로 입력되며, x축 값은 자동으로 0부터 1씩 증가하는 정수로 입력된다.

x축의 값을 생략할 경우 range(y축 데이터의 개수)로 표현할 수도 있습니다.

```python
# plot()함수에 두 개의 리스트 입력
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[12,43,25,15])
plt.show()
```

### 만약 x축 데이터와 y축 데이터의 개수가 맞지 않으면 에러가 발생한다.


```python
# plot()함수에 두 개의 리스트 입력
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[12,43,25])
plt.show()
```

### plot()함수에 두 개의 리스트를 입력했을 때는 첫 번째 리스트가 x축이고, 두 번째 리스트가 y축 값으로 입력된다

### plot함수에 기본 그래프를 그리기 위한 기본적인 세 단계
- import matplotlib.pyplot as plt : 라이브러리 불러오기
- plt.plot([x축 데이터], [y축 데이터]) : plot() 함수에 데이터 입력하기
- plt.show() : 그래프 보여주기

# 3. 그래프에 옵션 추가하기

### - 그래프에 제목 넣기

#### 제목을 넣는 함수 title('제목에 넣을 문자열') 

```python
import matplotlib.pyplot as plt
plt.title("plotting")
plt.plot([10,20,30,40])
plt.show()
```

### - 그래프에 범례 넣기

범례는 보통 두 개 이상의 데이터를 표시할 때 사용한다. 

label이라는 속성의 레이블 값으로 원하는 문자열을 넣어주고,

그래프를 그리기 전에 legend() 함수를 실행시키면 레이블 값이 범례로 나타난다.

두 가지 범례를 넣으려면 plot()함수를 두 개 사용하면 된다

```python
import matplotlib.pyplot as plt
plt.title("legend")
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend()
plt.show()
```

#### 범례의 위치를 직접 지정하려면

legend() 함수 안의 loc이라는 속성에 값을 넣어주면 됩니다.

이때 loc에 들어갈 숫자는 원하는 위치에 따라 0~10까지 넣을 수 있습니다.

```python
import matplotlib.pyplot as plt
plt.title("legend")
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend(loc = 4)
plt.show()
```

### - 그래프 색상 바꾸기

원하는 색상으로 바꾸려면 color 속성을 간단히 추가하면 됨.

color 속성을 생략하면 자동으로 색상이 설정된다.

```python
import matplotlib.pyplot as plt

plt.title('color') # 제목설정
# 그래프 그리기

plt.plot([10,20,30,40], color = 'r', label = 'skyblue')
plt.plot([40,30,20,10], color = 'pink', label = 'pink')
plt.legend() # 범례 표시
plt.show()
```

### 그래프 선 모양 바꾸기
linestyle 속성에 원하는 선 모양을 지정함.
linestyle 속성 대신 ls라고 작성가능

```python
import matplotlib.pyplot as plt
plt.title("linestyle") # 제목 설정
# 빨간색 dashed 그래프
plt.plot([10,20,30,40], color = 'r', linestyle = '--', label = 'dashed')
plt.plot([1,2,3,4],'b--', label = 'surprising!')

# 초록색 dotted 그래프
plt.plot([40,30,20,10], color = 'g', ls = ':', label = "dotted")
plt.legend() # 범례 표시
plt.show()
```

### 마커 모양 바꾸기
- plot()함수에 marker 속성을 설정하면 선 형태가 아닌 점 형태로 그래프를 그리기 가능
- 이때 색상과 마커(marker) 모양을 한번에 설정 가능
- 마커 모양 : 선, 점, 동그라미, 삼각형, 별 모양등 여러가지 

```python
import matplotlib.pyplot as plt
plt.title('marker') # 제목설정

# 빨간색 원형 마커 그래프
plt.plot([10,20,30,40], 'r:', label = 'circle') 

# 초록색 삼각형 마커 그래프
plt.plot([40,30,20,10], 'g^', label = 'triangle up')
plt.legend() # 범례 표시
plt.show()
```
