### 2. 데이터 리스트에 저장하기

```python
import csv
f = open('seoul.csv') # 맥이나 리눅스라면 f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data) # 맨 윗줄에 있는 헤더 부분을 제외
result = [] # 최고 기온 데이터를 담을 리스트

for row in data :
    if row[-1] != '' :  # 최고 기온 데이터 값이 존재한다면
        result.append(float(row[-1])) # result 리스트에 최고 기온 값 추가 (실수형)
print(result)
```

```python
print(len(result)) # 리스트에 데이터가 몇 개 인지
```

# 2. 데이터 시각화하기
- matplotlib 라이브러리 임포트
- plot() 함수에 데이터를 넣고
- show() 함수

```python
import matplotlib.pyplot as plt

plt.plot(result, 'r') # result 리스트에 저장된 값을 빨간색 그래프로 그리기
plt.figure(figsize=(12,12)) # 가로로 10인치, 세로로 2인치로 설정
plt.show()
```

#### * 그래프 크기 조정하기
- figure()함수의 figsize 속성 값을 변경하여 크기를 조절가능
- figsize = (가로 길이, 세로 길이) 형식으로 크기를 설정
- 단위는 인치 ( 1인치 = 2.54 센티미터)

#### 8월의 최고 기온 데이터 추출

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '':   # 최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '08':  # 8월에 해당하는 값이라면
            result.append(float(row[-1])) #  result 리스트에 최고 기온 값 추가
plt.plot(result, 'hotpink')   # result 리스트에 저장된 값을 hotpink 색으로 그리기
plt.show()
```

#### 매년 2월 14일의 최고 기온 데이터 추출

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '':   # 최고 기온 값이 존재한다면
        if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14':  # 2월 14일에 해당하는 값이라면
            result.append(float(row[-1])) #  result 리스트에 최고 기온 값 추가
plt.title('해가 흐를수록 최고 기온이 증가하는 추세')
plt.plot(result, 'hotpink')   # result 리스트에 저장된 값을 hotpink 색으로 그리기
plt.show()
```

#### 1983년 이후 데이터만 추출해서 최고 기온 및 최저 기온 시각화

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = [] # 최고 기온 값을 저장할 리스트 생성
low = [] # 최저 기온 값을 저장할 리스트 생성



for row in data:
    if row[-1] != '' and row[-2] != '' :  # 최고 기온 값과 최저 기온 값이 존재한다면
        if 1983 <= int(row[0].split('-')[0]) : # 1983년 이후 데이터라면
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '14': # 2월 14일이라면
                high.append(float(row[-1])) # 최고 기온 값을 high list에 저장
                low.append(float(row[-2])) # 최저 기온 값을 low list에 저장

                
                
plt.title('내 생일의 기온 변화 그래프') # 제목
plt.rc('font', family = 'Malgun Gothic') # 맑은 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지

plt.plot(high, 'hotpink', label = '최고 기온') # high 리스트에 저장된 값을 hotpink 색으로 그리기
plt.plot(low, 'skyblue', label = '최저 기온') # low 리스트에 저장된 값을 skyblue 색으로 그리기
plt.legend() # 범례 표시
plt.show() # 그래프 나타내기



```
