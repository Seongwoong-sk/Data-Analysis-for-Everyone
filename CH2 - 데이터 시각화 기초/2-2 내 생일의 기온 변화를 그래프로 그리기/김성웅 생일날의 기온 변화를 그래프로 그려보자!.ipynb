```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

high = []
low = []

for row in data :
    if row[-1] != '' and row[-2] !='' :  # 최고기온과 최저기온이 존재한다면
        if 1910 <= int(row[0].split('-')[0]):
            if row[0].split('-')[1] == '11' and row[0].split('-')[2] == '10':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font', family = 'Malgun Gothic')
plt.title("김성웅 생일날 년도의 변화")
plt.rcParams['axes.unicode_minus'] = False 
plt.plot(high, 'r' , label = '최고 기온')
plt.plot(low, 'b' , label = '최저 기온')
plt.legend()
plt.show()
```
