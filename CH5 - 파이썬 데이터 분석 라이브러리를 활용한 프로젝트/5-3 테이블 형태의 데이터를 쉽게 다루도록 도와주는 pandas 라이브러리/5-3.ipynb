### 2차원 배열 형태의 데이터 프레임 연산
- 행을 기준으로 계산할 수도 있고, 열을 기준으로 계산할 수도 있음.
- 행과 열에 대한 연산을 동시에 할 때, 행 우선 계산과 열 우선 계산을 구분하여 생각해 볼 수 있다.

- 2차원 데이터일 경우 pandas에서는 기본적으로 행 방향을 축으로 계산한다. 하지만 경우에 따라 열 방향을 축으로 계산하기도 한다.
    - 이때 열 방향을 축으로 계산하려면 axis = 1, axis = 0은 행 방향을 나타내며 0이 기본 값이다.

#### 행 방향 축을 기준으로 한 연산
- 전체 행에 대해서 A열의 값을 B열의 값으로 나눈 후, 그 결과를 새로 만든 D열에 저장

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B'] # A열의 값을 B열의 값으로 나눈 값을 D열에 저장
df
```

#### 열 방향 축을 기준으로 한 연산
* df 프레임에서 행 우선으로 합을 구하고, 그 결과를 E라는 열을 생성해서 저장

```python
df['E'] = np.sum(df, axis= 1) # 행 우선 계산 값을 E열에 저장
df.head()
```

#### 전체 데이터에 대해서 열 우선 계산을 하는 방법
- 1개의 열을 1개의 열로 나는 계산은 산술 연산자로 가능
- 여러 개의 열에 대한 계산을 하려면 pandas 라이브러리를 사용해야 함

```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B'] 
df['E'] = np.sum(df, axis = 1)
df = df.sub(df['A'], axis = 0) # A열의 데이터를 기준으로 열 우선 계산
df.head()
```

```python
# div 함수 사용

import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)
df = pd.DataFrame(np.random.rand(8,3), index = index, columns = list('ABC'))
df['D'] = df['A'] / df['B'] 
df['E'] = np.sum(df, axis = 1)
df = df.sub(df['A'], axis = 0) # A열의 데이터를 기준으로 열 우선 계산
df = df.div(df['C'], axis = 0) # C열 데이터를 기준으로 열 우선 계산
df.to_csv('test.csv') # 데이터 프레임을 test.csv 파일로 저장
df.head()
```

# 4. pandas로 인구 구조 분석하기

Unit 14에서 파이썬 기본 문법과 numpy 라이브러리를 활용해서 만들었던 데이터 프로젝트를 pandas 라이브러리로 구현하기

## 알고리즘 
* 1. 데이터를 읽어온다
    - 전체 데이터를 총 인구수로 나누어 비율로 변환한다.
    - 총인구수와 연령구간인구수를 삭제한다.
* 2. 궁금한 지역의 이름을 입력받는다.
* 3. 궁금한 지역의 인구 구조를 저장한다.
* 4. 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역을 찾는다.
    - [1]. 전국의 모든 지역 중 한 곳(B)을 선택한다.
    - [2]. 궁금한 지역 A의 0세 인구 비율에서 B의 0세 인구 비율을 뺀다.
    - [3]. [2]를 100세 이상 인구수에 해당하는 값까지 반복한 후 차이의 제곱을 모두 더한다.
    - [4]. 전국의 모든 지역에 대해 반복하며 그 차이가 가장 작은 지역을 찾는다.
* 5. 가장 비슷한 곳의 인구 구조와 궁금한 지역의 인구 구조를 시각화한다.


### 1. 데이터 읽어오기 

```python
import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col = 0)
df.head()
```

#### 연령별 인구 숫자를 비율로 바꿔주고 총인구수와 연령구간인구수 데이터를 삭제하기 (del 명령 사용)

```python
import pandas as pd
df = pd.read_csv('age.csv', encoding='cp949', index_col = 0)
df= df.div(df['총인구수'], axis = 0) # 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수'] # 총인구수, 연령구간인구수 열 삭제
```

### 2~3 궁금한 지역 이름 입력받고 해당 지역의 인구 구조 저장하기
- df.index.str.contains() 함수는 데이터 프레임의 인덱스 문자열에 원하는 문자열이 포함된 행을 찾아냄

```python
name = input('원하는 지역의 이름을 입력해주세요 : ')  # 2 지역 이름 입력
a = df.index.str.contains(name)
df2 = df[a]
df2
```

 데이터를 그래프로 그리려면 행과 열을 바꾼후 plot() 함수 실행 : 인덱스가 가로축이 되기 때문

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
plt.show()
```

### 4~5 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 시각화하기
- 궁금한 지역의 연령별 비율과 다른 지역의 연령별 비율의 차이를 계산

```python
import numpy as np

# 4-1 궁금한 지역 A의 인구 비율에서 B의 인구 비율을 뺀다
x = df.sub(df2.iloc[0], axis = 1)

# 4-2 A의 인구 비율에서 B의 인구 비율을 뺀 값의 제곱 값을 모두 더한다.
y = np.power(x, 2)
z = y.sum(axis = 1)
```

* pandas의 sort_values() 정렬 함수와 슬라이싱을 이용하면 상위 몇 개 지역까지 쉽게 찾을 수 있음

```python
i = z.sort_values().index[:5] # 4-3 그 차이가 가장 작은 지역 5곳을 찾는다.
df.loc[i].T.plot() # 4-4 결과를 꺾은선 그래프로 보여준다
plt.show()
```

```python
# 4 - 궁금한 지역의 인구 구조와 가장 비슷한 인구 구조를 가진 지역 찾기
# 위의 코드를 한 줄로

df.loc[np.power(df.sub(df2.iloc[0], axis = 1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
```

# 전체 코드

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')
df = pd.read_csv('age.csv', encoding = 'cp949', index_col=0)
df = df.div(df['총인구수'], axis = 0)
del df['총인구수'], df['연령구간인구수']

name = input('원하는 지역의 이름을 입력해주세요 : ')
a = df.index.str.contains(name)
df2 = df[a]


df.loc[np.power(df.sub(df2.iloc[0], axis= 1), 2).sum(axis= 1).sort_values().index[:5]].T.plot()

plt.show()
```
