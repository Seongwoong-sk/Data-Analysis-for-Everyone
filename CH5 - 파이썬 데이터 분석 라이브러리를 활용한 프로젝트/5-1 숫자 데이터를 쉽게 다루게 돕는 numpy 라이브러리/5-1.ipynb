### * 버블차트 그리기 (일반코드 vs numpy 코드)

```python
# 일반코드

import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(200):
    x.append(random.randint(10,100))
    y.append(random.randint(10,100))
    size.append(random.randint(10,100))
    
plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

```python
# numpy를 이용한 코드

import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(10,100,200)  # 10 ~ 100 사이의 숫자를 200번 랜덤으로 추출
y = np.random.randint(10,100,200)
size = np.random.rand(100) * 100  # 0 ~ 1 사이에 있는 n개의 실수(float)을 만들어라.

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```

* np.random.rand(n) = 0 ~ 1 사이에 있는 n개의 실수를 만드는 함수
* np.random.randint(a,b,n) = a 이상 b 이하인 정수 n개를 만드는 함수

# 3. numpy array 생성하기

```python
#  리스트와 같은 연속된 데이터는 numpy array로 표현할 수 있다.
import numpy as np
a = np.array([1,2,3,4])
print(a)
```

* numpy array는 리스트 내 요소를 구별하는 쉼표(,)과 없다
* 리스트의 특징인 인덱싱과 슬라이싱을 할 수 있다

```python
a = np.array([1,2,3,4])
print(a[1], a[-1]) # a의 1번 인덱스 값, -1번 인덱스 값 출력
print(a[1:]) # a의 1번 인덱스를 기준으로 슬라이싱 결과 출력
```

* numpy array에는 한 가지 데이터 타입의 데이터만을 저장할 수 있다. (숫자와 문자가 함께 저장되면 문자로 변환되어 저장)
* 리스트는 정수, 문자열, 리스트 등 다양한 데이터 타입을 담을 수 있다.

```python
a = np.array([1,2,'3',4])
print(a)
```

* 배열을 생성하기 위해 꼭 리스트가 있어야 하는 것은 아니다.
* zeros() , ones() ,eye() 같은 다양한 함수를 사용해서 배열을 초기화 할 수 있음
* 이 세가지 함수는 0 또는 1로 배열을 채우고 싶을 때 매우 유용하다.

```python
import numpy as np
a = np.zeros(10) # 0으로 이루어진 크기가 10인 배열 생성
print(a)
```

```python
a = np.ones(10) # 1로 이루어진 크기가 10인 배열 생성
print(a)
```

```python
a = np.eye(3) # 3행 x 3열의 단위 배열 생성
print(a)
```

* 0과 1이 아닌 연속된 숫자로 데이터를 생성하기

```python
print(np.arange(3)) # arange() 함수에 1개 값 입력 : 0부터 해당 숫자보다 1만큼 작은 정수까지 저장된 배열이 생성
print(np.arange(3,7)) # arange() 함수에 2개 값 입력 : 첫번째 숫자부터 두번째 숫자보다 1만큼 작은 정수까지 저장된 배열 생성
print(np.arange(3,7,2)) # arange() 함수에 3개 값 입력 : 첫번째 숫자부터 두번째 숫자보다 1만큼 작은 정수까지의 범위에서 
                        # 세 번째 숫자만큼의 간격을 둔 숫자가 저장된 배열이 생성
```

```python
# numpy의 linspace() 함수를 사용가능

import numpy as np
a = np.arange(1,2,0.1) # 1이상 2미만 구간에서 0.1 간격으로 실수 생성
b = np.linspace(1,2,11) # 1부터 2까지 11개 구간으로 나눈 실수 생성
print(a)
print(b)
```

* np.arange() : 간격을 지정함 / 특정 간격에 해당하는 값을 생성하고 싶을 때
* np.linspace() : 구간의 개수를 지정함 / 특정 개수의 구간으로 나눈 값을 생성하고 싶을 때

```python
a = np.arange(-np.pi, np.pi, np.pi/10)
b = np.linspace(-np.pi, np.pi,20)
print('a : ',a)
print('b : ',b)
```

### mask 기능 : 어떤 조건에 부합하는 데이터만 선별적으로 저장하기 위한 기능

```python
a = np.arange(-5,5)
print(a)
```

```python
# 조건에 맞으면 True, 반대는 False인 마스크 
print(a<0)
```

```python
# 마스크를 배열에 적용하면 조건에 부합하는 데이터만 남음

print(a[a<0])

# 변수에 넣어서 사용 가능
mask1 = abs(a) > 3
print(a[mask1])

# 여러 개의 마스크를 연결하여 사용 가능
mask2 = abs(a) > 3
mask3 = abs(a) % 2 == 0
print(a[mask2+mask3]) # 둘 중 하나의 조건이라도 참일 경우
print(a[mask2*mask3]) # 두 가지 조건이 모두 참일 경우
```

### numpy 라이브러리를 사용하여 재미있는 버블 차트 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000) # 1000개의 랜덤 추출
y = np.random.randint(-100, 100, 1000) # 100개의 랜덤 추출
size = np.random.rand(100) * 100

mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1+mask2]
y = y[mask1+mask2] # mask1과 mask2 중 하나라도 만족하는 값 저장

plt.scatter(x, y, s= size, c=x, cmap = 'jet', alpha=0.7)
plt.colorbar()
plt.show()
```
